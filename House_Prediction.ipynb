## Independent and Dependent Features

```python
X=df.iloc[:,:-1] #skip last column
y=df.iloc[:,-1] #All datapoints in last column
```

```python
X.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

## Standardize the Dataset (same scale)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test) # does not fit_transform because we are going to make sure that whatever
                               # information i have with respect to the training dataset and whatever techniques i have applied
                               # for transforming it ... the same techniques needs to be applied to the test data
```

```python
X_train
```

```python
X_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
#In order to train model:
regression.fit(X_train,y_train)
```

## Print the Coefficients and the Intercept

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameters the model has been trained
regression.get_params()
```

## Prediction with Test Data

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
# Assumptions

## Plot scatter plot for prediction

plt.scatter(y_test,reg_pred)                  # y_test is actual data
plt.show()
```

###### Here plotting of values is basically linear. So when this plotting is linear , your model is performing well. y_test is actual value and reg_pred is predicted value. When this both values are almost linear like this, then your model are actually done and it is predicting well

### Plot with respect to the Residuals

###### Residuals is nothing but the errors.  Error with respect to y_test and regression pred

```python
residuals = y_test - reg_pred
```

```python
residuals
```

```python
#Plot this residuals
sns.displot(residuals,kind="kde")
plt.show()
```

###### Here maximum difference is between -10 to +10 and some points are between 10 to 30 which are outliers.

```python
## Scatter plot with respect to prediction and residuals
plt.scatter(reg_pred,residuals)
plt.show()
#uniform distribution
```

## Performance Metrix

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

#### R square and adjusted R square


#### Formula
#### R^2 = 1-SSR/SST        SSR-sum of square residuals , SST - sum of square total

#### Adjusted R^2 = 1-[(1-R^2)*(n-1)/(n-k-1)]   

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
#Adjusted R-squared
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
